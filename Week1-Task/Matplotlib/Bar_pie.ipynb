# Bar Chart and pie chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["Jan","Feb","Mar","Apr","May","Jun"])
y = np.array([5000,1000,3000,4500,500,7000])

plt.bar(x,y)
plt.title("Month wise Growth")
plt.xlabel("Month")
plt.ylabel("Subscriber Increase")
```

```python
x = np.array(["Long Form","Shorts","live","Post"])
y = np.array([4060,1200,300,500])
plt.pie(y,labels=x,autopct="%1.1f%%",colors=["#1F77B4","#FF7F0E","#2CA02C","#9467BD"],shadow=True,explode=[0,0.1,0,0.2])
plt.title("Subcriber Details")
```
