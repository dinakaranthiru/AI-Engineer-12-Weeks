# Scatter Plot & Histogram 


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12,10,11,14,7,6,10,14])
y = np.array([5000,4000,3000,7000,6000,1000,1500,9000])
x1 = np.array([15,13,14,17,10,9,13,17])
y1 = np.array([8000,7000,6000,11000,9000,4000,4500,13000])
plt.scatter(x,y,color="red",alpha=1,edgecolors="blue",label="Logic")
plt.scatter(x1,y1,color='blue',alpha=0.5,label="code io")
plt.title("Scatter Plot")
plt.legend()
```

```python
x = np.array([10,12,14,9,12,13,5,2,21])
plt.hist(x,bins=5)
```
