# Scatter Plot & Histogram 


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12,10,11,14,7,6,10,14])
y = np.array([5000,4000,3000,7000,6000,1000,1500,9000])
x1 = np.array([15,13,14,17,10,9,13,17])
y1 = np.array([8000,7000,6000,11000,9000,4000,4500,13000])
plt.scatter(x,y,color="red",alpha=1,edgecolors="blue",label="Logic")
plt.scatter(x1,y1,color='blue',alpha=0.5,label="code io")
plt.title("Scatter Plot")
plt.legend()
```

```python
x = np.array([10,12,14,9,12,13,5,2,21])
plt.hist(x,bins=5)
```

```python
print(plt.subplots(2,2))
```

```python
x = np.array([2021,2022,2023,2024,2025,2026])
y = np.array([0,10000,60000,100000,200000,300000])

#figure combined whole image
#indidual subplot
figure,axes = plt.subplots(2,2)

axes[0,0].plot(x,y,c="red")
axes[1,0].scatter(x,y,c="blue")
axes[0,1].plot(x,y,c="green")
axes[1,1].scatter(x,y,c="yellow")
```
