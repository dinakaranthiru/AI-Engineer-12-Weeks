```python
x = [2021,2022,2023,2024,2025,2026]
y = [0,10000,60000,80000,100000,200000]
plt.plot(x,y,marker=".",markeredgewidth="1",markerfacecolor="red",markersize='10')
plt.title("Subcriber's Count")
plt.xlabel('Year')
plt.ylabel('Subcribers')
plt.subplots_adjust(left=0.1)
plt.tick_params(axis="x",labelcolor="red")
plt.tick_params(axis="y",labelcolor="blue")
plt.grid(color='green',linestyle='dashed')
plt.show()
```
