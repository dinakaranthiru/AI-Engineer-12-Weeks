# Week 1 — Python for Data Analysis
## Dataset: Titanic Passenger Data
**Objective:** Load the Titanic dataset, perform data exploration, detect missing values,
analyze data types, and visualize key patterns.

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
print("Libraries loaded successfully!")
```

## Section 1: Load the Dataset

```python
# Load Titanic dataset directly via Seaborn (no file download needed)
df = sns.load_dataset("titanic")

# Alternatively, if using a CSV file:
# df = pd.read_csv("titanic.csv")

# Confirm it loaded
print("Dataset loaded!")
print(f"Shape: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
df
```

## Section 2: Data Summary / Overview


```python
# ── Basic shape and columns ──────────────────────────────────────────
print("=" * 50)
print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")
print("=" * 50)

# ── Column names ─────────────────────────────────────────────────────
print("\nColumn Names:")
print(df.columns.tolist())

# ── First and last 5 rows ─────────────────────────────────────────────
print("\nFirst 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())

```

```python
# ── Statistical summary for numeric columns ───────────────────────────
print("\nDescriptive Statistics (Numeric Columns):")
display(df.describe())

# ── Statistical summary for categorical columns ───────────────────────
print("\nDescriptive Statistics (Categorical Columns):")
display(df.describe(include=["object", "category"]))
```

## Section 3: Missing Value Detection


```python
# ── Count of missing values per column ───────────────────────────────
print("Missing Values per Column:")
missing_count = df.isna().sum()
print(missing_count)

# ── Percentage of missing values ──────────────────────────────────────
print("\nPercentage of Missing Values (%):")
missing_percent = (df.isna().mean() * 100).round(2)
print(missing_percent)

# ── Summary table: only columns WITH missing values ───────────────────
missing_summary = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing %": missing_percent
}).query("`Missing Count` > 0").sort_values("Missing %", ascending=False)

print("\nColumns with Missing Values:")
display(missing_summary)
```

```python
# ── Visualize missing values as a heatmap ────────────────────────────
plt.figure(figsize=(12, 4))
sns.heatmap(df.isna(), cbar=False, cmap="viridis", yticklabels=False)
plt.title("Missing Values Heatmap (Yellow = Missing)", fontsize=14)
plt.tight_layout()
plt.show()
```

## Section 4: Data Type Analysis


```python
# ── Check data types ─────────────────────────────────────────────────
print("Data Types of Each Column:")
print(df.dtypes)

# ── Separate numeric and categorical columns ──────────────────────────
numeric_cols    = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object", "category", "bool"]).columns.tolist()

print(f"\nNumeric Columns    ({len(numeric_cols)})  : {numeric_cols}")
print(f"Categorical Columns ({len(categorical_cols)}) : {categorical_cols}")
```

```python
# ── Value counts for key categorical columns ──────────────────────────
for col in ["sex", "pclass", "embarked"]:
    print(f"\nValue counts — '{col}':")
    print(df[col].value_counts())
```

## Section 5: Data Visualization

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["age"].dropna(), kde=True, bins=30, color="steelblue")
plt.title("Age Distribution of Passengers", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="survived", data=df, palette="Set2")
plt.title("Survival Count (0 = Died, 1 = Survived)", fontsize=14)
plt.xlabel("Survived")
plt.ylabel("Count")
plt.xticks([0, 1], ["Died", "Survived"])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="sex", hue="survived", data=df, palette="Set1")
plt.title("Survival Count by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["Died", "Survived"])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="pclass", hue="survived", data=df, palette="coolwarm")
plt.title("Survival Count by Passenger Class", fontsize=14)
plt.xlabel("Passenger Class (1 = First, 3 = Third)")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["Died", "Survived"])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="age", y="fare", hue="survived", data=df, palette="Set1", alpha=1)
plt.title("Age vs Fare (Colored by Survival)", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
corr = df.select_dtypes(include=["int64", "float64"]).corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap (Numeric Columns)", fontsize=14)
plt.tight_layout()
plt.show()
```

## Section 6: Key Insights & Conclusion

### Summary of Findings:
- **Dataset:** 891 rows, 15 columns — Titanic passenger data
- **Missing Data:** `age` (~20% missing), `deck` (~77% missing — mostly unusable)
- **Data Types:** Mix of numeric (age, fare) and categorical (sex, pclass, embarked)
- **Survival Rate:** Only ~38% of passengers survived
- **Gender Impact:** Females survived at a much higher rate than males
- **Class Impact:** 1st class passengers had the best survival rate
- **Age & Fare:** Younger passengers and higher-fare payers had slightly better survival odds
