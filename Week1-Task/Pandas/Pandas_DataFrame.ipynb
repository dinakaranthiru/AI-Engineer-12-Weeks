# Dataframe in pandas
In pandas, a DataFrame is:

A 2‑dimensional labeled table (rows × columns).
Each column is a Series (so a DataFrame is like a collection of Series sharing the same row index).
Columns can have different data types (int, float, string, datetime, etc.).
Very similar to:
A table in a database,
An Excel sheet, or
A CSV file loaded into memory.

```python
import pandas as pd
```

```python
student = {
    "Name":["Messi","Neymar","Lamine","Gavi","Pedri"],
    "Age":[38,32,18,22,24],
    "CGPA":[9.9,8.7,8.0,9.1,9.2]
}
df = pd.DataFrame(student)
df
```

Giving index of our own

```python
df = pd.DataFrame(student,index=["Std1","Std2","Std3","Std4","Std5"])
df
```

Accessing element using index

```python
print(df.loc["Std1"])
print(df.iloc[1])
```

To add a new column

```python
df["Medium"] = ["English","English","Tamil","Tamil","Tamil"]
```

```python
df
```

To add a new row

```python
std4 = pd.DataFrame({"Name":"RashFord","Age":28,"CGPA":8.8,"Medium":"CBSE"},index=["Std 6"])
```

```python
df = pd.concat([df,std4])
df
```

TO add Multiple rows

```python
student = pd.DataFrame([
    {"Name":"Dani Olama", "Age":28,"CGPA":8.7,"Medium":"English"},
    {"Name":"Curbasi","Age":22,"CGPA":8.9,"Medium":"English"}
],index=["Std7","Std8"])
```

```python
df = pd.concat([df,student])
df
```

```python
# Remove duplicate rows based on all columns, keep first occurrence
df = df.drop_duplicates(keep="first")
# Filter out rows matching that specific record
df = df[~((df["Name"] == "Dani Olama") &
          (df["Age"] == 28) &
          (df["CGPA"] == 8.8))]
df
```

#  Small End-to-End Analysis

```python
import pandas as pd

# 1) Load data (assume CSV of monthly sales)
df = pd.read_csv("monthly_sales.csv")
# Columns: month, region, product, units, price_per_unit

# 2) Create revenue column
df["revenue"] = df["units"] * df["price_per_unit"]

# 3) Basic checks
print(df.head())
print(df.describe())

# 4) Total revenue by region
rev_by_region = df.groupby("region")["revenue"].sum()

# 5) Top 5 months by revenue
top_months = df.sort_values(by="revenue", ascending=False).head(5)

print("Revenue by region:")
print(rev_by_region)

print("Top 5 month-level rows by revenue:")
print(top_months)
```
