# Data Cleaning using pandas

```python
import pandas as pd
```

```python
df = pd.read_json('sample1.json')
df.head()
```

# dropping a cloumn

```python
df = df.drop(columns=["id"])
df.head()
```

# To clean NaN in data

```python
df = df.dropna(subset=["marks"])
df
```

# if we have an NaN in an column instead of NaN we can define what value need to present in the NaN position

```python
df = df.fillna({
    "marks":99
})
df
```

# To replace a value

```python
df['class'] = df['class'].replace({
    "Class A": "IT",
    "Class B": "CSE",
    "Class C": "EEE"
})
df
```

# To remove duplicates

```python
df = df.drop_duplicates()
df
```
