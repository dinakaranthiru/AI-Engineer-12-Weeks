# Data cleaning = the process of fixing/Removing:
iincomplete, incorrect or irrelevant data.
~75% of work done with pandas is data cleaning.

```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset_Sample.csv')
df.head()
```

# Drop irrelevant columns


```python
df = df.drop(columns=['Legendary','No']) #removing the column legendary and no form the dataset
df
```

# HAndling missing data

```python
# df = df.dropna(subset=['Type2'])
df = df.fillna({"Type2":"None"}) #making the type2 column missing value into NONE instead of Nan is replace it with None
df
```

# Fix inconsistent values

```python
df['Type1'] = df['Type1'].replace({'Grass':'GRASS','Fire':'FIRE','Water':'WATER'}) # replace the values in the type1 column into upper case
df.head(50)
```

# Standardizre text

```python
df['Name'] = df['Name'].str.lower() #making all the name in column name into smaller case
df
```

# Fix data types

```python
df['Legendary'] = df['Legendary'].astype(bool) # previously the data in the legendary column is 0 or 1 now we changed the data type to boolean so it now in 0 For False 1 For True
df.tail(60)
```

# Remove duplicate values

```python
df = df.drop_duplicates()
df
```
