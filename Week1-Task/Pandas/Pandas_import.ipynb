# Aggregate in Pandas

```python
df = pd.read_json('sample1.json')
df
```

```python
print(df.mean(numeric_only=True)) #it only take average of only numeric values
print(df.max(numeric_only=True))
print(df["marks"].min()) #we are specifying a particular column for min
print(df["marks"].sum())
```

# Group by


```python
group = df.groupby("marks")
group
# print(group.mean(numeric_only=True))
print(group["marks"].sum())
```
