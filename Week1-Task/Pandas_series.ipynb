# Series in pandas
In pandas, a Series is:
A one‑dimensional labeled array.
Holds data of one type (int, float, string, etc.).
Think of it as:
A single column in Excel, with row labels.

```python
age = [21,22,23,24,25]
series = pd.Series(age)
series
```

using index to acces element


```python
print(series.loc[1])
```

Creating our own index insted of 0 1 2 3 4


```python
age = [21,22,33,44,55]
series = pd.Series(age,index=["Gavi","Pedri","Lamine","Raphina","Dani"])
series
```

accessing element using the index:
-- loc
-- iloc

```python
age = [21,22,33,44,55]
series = pd.Series(age,index=["Gavi","Pedri","Lamine","Raphina","Dani"])
series
print(series.loc["Gavi"])
# print(series.loc[0]) -- this one will show error
print(series.iloc[0])
print(series[series>22])
```

Using Dictionary 

```python
age = {
    "Gavi":23,
    "Messi":38,
    "Julian":25,
    "Ronaldo":39
}
series = pd.Series(age)
print(series[series>30])
```
